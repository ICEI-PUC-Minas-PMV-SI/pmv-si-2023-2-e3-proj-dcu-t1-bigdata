```python
# função criada para remover os caracteres especiais para a padronização

def remover_acentos_e_cedilha2(texto):
    # Mapeamento de caracteres com acentos e cedilha
    mapeamento = {
        'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u',
        'à': 'a', 'è': 'e', 'ì': 'i', 'ò': 'o', 'ù': 'u',
        'â': 'a', 'ê': 'e', 'î': 'i', 'ô': 'o', 'û': 'u',
        'ã': 'a', 'õ': 'o', 'ç': 'c', 'Á': 'A', 'É': 'E',
        'Í': 'I', 'Ó': 'O', 'Ú': 'U', 'À': 'A', 'È': 'E',
        'Ì': 'I', 'Ò': 'O', 'Ù': 'U', 'Â': 'A', 'Ê': 'E',
        'Î': 'I', 'Ô': 'O', 'Û': 'U', 'Ã': 'A', 'Õ': 'O',
        'Ç': 'C'
    }
    # Substituir caracteres no texto
    for acento, sem_acento in mapeamento.items():
        texto = texto.replace(acento, sem_acento)
    return texto
```
